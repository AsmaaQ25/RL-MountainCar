```python
# import needed libraries
import gym
from stable_baselines3 import DDPG
from stable_baselines3.ddpg.policies import MlpPolicy
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common import results_plotter
from stable_baselines3.common.callbacks import CheckpointCallback, EvalCallback

from stable_baselines3.common.noise import  OrnsteinUhlenbeckActionNoise
import numpy as np
import os
```

```python
# Create log dir
log_dir = "tmp/gym/"
os.makedirs(log_dir, exist_ok=True)

# Create and wrap the environment
env = gym.make("MountainCarContinuous-v0")

# Logs will be saved in log_dir/monitor.csv
env = Monitor(env, log_dir)
```

------------------------------------

```python
%%time

#create action noise 
n_actions = env.action_space.shape[-1]
action_noise = OrnsteinUhlenbeckActionNoise(mean=np.zeros(n_actions), sigma=float(0.5) * np.ones(n_actions))
model = DDPG("MlpPolicy", env, action_noise=action_noise, verbose=1)

# Save a checkpoint every 1000 steps
checkpoint_callback = CheckpointCallback(save_freq=1000, save_path=log_dir,
                                         name_prefix='rl_model')
# another callback method, also Save a checkpoint every 1000 steps
eval_callback = EvalCallback(env, best_model_save_path=log_dir,
                             log_path=log_dir, eval_freq=1000,
                             deterministic=True, render=False)

#train the model
model.learn(total_timesteps=300000, callback= checkpoint_callback)

# now save the model
model.save("t4_ddpg_mountain")
```

-------------------------------

```python
# load the model to skip retraining
model = DDPG.load("t4_ddpg_mountain")
```

```python
# Now plot the rewards saved in monitor.csv
results_plotter.plot_results([log_dir], 
                             1e5,
                             results_plotter.X_TIMESTEPS,
                             "Mountain Car learning progress")
```

```python
# now lets see the agent after learning
''' For this to work and to see the car going uphill in the first try
    you need to either load the model or retrain it'''
#reset environment
state = env.reset()

import time
flag = True

# give it 100 steps to reach terminal state -uphill-
for t in range(100):
    
    action, _ = model.predict(state)
    next_state, reward, done, info = env.step(action)
    if done != True:
        state = next_state
        env.render()
    # this part is to slow down the motion for visibility
    if flag:
        time.sleep(4)
        flag = False
    time.sleep(0.1)
    # close at terminal state
    if done:
        time.sleep(3)
        env.reset()
        env.close()
        break
```
